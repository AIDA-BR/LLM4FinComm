```python
import pandas as pd
import fitz
```

```python
file_input = f""
file_output = f""
```

```python
def extract_text_from_pdf(pdf_path):
    try:
        document = fitz.open(pdf_path)
        text = ""
        for page in document: # Catch text through all pages
            text += page.get_text()
        return text
    except Exception as e:
        print(f"Erro{pdf_path}: {e}")
        return ""
```

```python
pdf_texts = []
df = pd.read_csv(file_input)
for index, row in df.iterrows():
    pdf_filename = row['fr_pdf']
    pdf_path = f"fatos_relevantes/{pdf_filename}"
    pdf_text = extract_text_from_pdf(pdf_path)
    pdf_texts.append(pdf_text)

df['pdf_text'] = pdf_texts
df.to_csv(file_output, index=False)
```
